# How to calculate Zwicker loudness from .wav file (time varying signal)

MoSQITo is a unified and modular development framework of key sound quality metrics favoring reproducible science and efficient shared scripting among engineers, teachers and researchers community.

This tutorial explains how to calculate the acoustic loudness (Zwicker method) of a time-varying signal from a .wav file.


```python
# Add MoSQITo to the Python path
import sys
sys.path.append('..')

# Useful package import
import IPython

#local applications imports
from mosqito.Classes.Audio_signal import Audio_signal
```

For this tutorial, the test signal nÂ°24 from ISO 532-1 annex B5 is used. It is the recording of a woodpecker that can be heard by using the command below.
According to the standard, for the calibration of the WAVE file, 0 dB (relative to full scale) should correspond to a sound pressure level of 100 dB.

```python
IPython.display.Audio("../mosqito/tests/data/ISO_532-1/Annex B.5/Test signal 24 (woodpecker).wav")
```

The audio signal object to be analyzed is created according to the "Audio_signal" class, then the signal from the .wav file is loaded using the "load_wav" method which extracts its signal values and sampling frequency from 2 entries : a "is_stationary" boolean and the file path.

```python
audio = Audio_signal()
audio.load_wav(False,"../mosqito/tests/data/ISO_532-1/Annex B.5/Test signal 24 (woodpecker).wav")
```

Then the signal is filtered by an array of 28 one-third-octave-band filters, from 25 Hz to 12 500 Hz. It is then squared and smoothed by third-order low-pass-filters. This is done by using the "comp_third_oct" method.

```python
audio.comp_third_oct()
```

The loudness calculation is done using the "comp_loudness" method. The script calculates the loudness N in sones versus time, the specific loudness N_specific in sones/Bark versus time and the corresponding critical band rate scale bark_axis in Bark.

The calculation being done, the results can be plotted or used for further analysis.

```python
audio.comp_loudness()
audio.plot_loudness()    
```
